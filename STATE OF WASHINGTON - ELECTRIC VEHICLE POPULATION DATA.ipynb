# COLUMNS CONTENTS

## VIN (1-10)

The 1st 10 characters of each vehicle's Vehicle Identification Number (VIN).

## Vehicle Location

The center of the ZIP Code for the registered vehicle.
geocoded_column 

## State

This is the geographic region of the country associated with the record. These addresses may be located in other states.

## Postal Code

The 5 digit zip code in which the registered owner resides.

## Model Year

The model year of the vehicle, determined by decoding the Vehicle Identification Number (VIN).

## Model

The model of the vehicle, determined by decoding the Vehicle Identification Number (VIN).

## Make

The manufacturer of the vehicle, determined by decoding the Vehicle Identification Number (VIN).

## Legislative District

The specific section of Washington State that the vehicle's owner resides in, as represented in the state legislature.

## Electric Vehicle Type

This distinguishes the vehicle as all electric or a plug-in hybrid.

## Electric Utility

This is the electric power retail service territories serving the address of the registered vehicle. All ownership types for areas in Washington are included: federal, investor owned, municipal, political subdivision, and cooperative. If the address for the registered vehicle falls into an area with overlapping electric power retail service territories then a single pipe | delimits utilities of same TYPE and a double pipe || delimits utilities of different types. 

## Electric Range

Describes how far a vehicle can travel purely on its electric charge.

## DOL Vehicle ID

Unique number assigned to each vehicle by Department of Licensing for identification purposes.

## County

This is the geographic region of a state that a vehicle's owner is listed to reside within. Vehicles registered in Washington state may be located in other states.

## Clean Alternative Fuel Vehicle (CAFV) Eligibility

This categorizes a vehicle as a Clean Alternative Fuel Vehicle (CAFV) based on the fuel and electric-only range requirements in House Bill 2042 as passed during the 2019 legislative session.

## City

The city in which the registered owner resides.

## Base MSRP

This is the lowest Manufacturer's Suggested Retail Price (MSRP) for any trim level of the model in question.

## 2020 Census Tract

The census tract identifier is a combination of the state, county, and census tract codes as assigned by the United States Census Bureau in the 2020 census, also known as Geographic Identifier (GEOID).

## DATA TYPE CONVERSITIONS

```python
df['Model Year'] =pd.to_datetime(df['Model Year'],format='%Y')   
```

### CHECKING MISSING VALUES

```python
Missing_values = df.isnull().sum()
print(Missing_values)
```

### HANDLING MISSING VALUES

```python
df.loc[df['County'].isna(),'County'] = 'Unknown'
```

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
df['Postal Code'] = df['Postal Code'].fillna(df['Postal Code'].mean())
```

```python
df['Electric Range'] = df['Electric Range'].fillna(df['Electric Range'].median()) # used median because outliers are there
```

```python
df['Base MSRP'] = df['Base MSRP'].fillna(df['Base MSRP'].median()) # used median because outliers are there
```

```python
df['Legislative District'] = df['Legislative District'].fillna('0')
```

```python
df = df.dropna(subset=['Vehicle Location']) # droped null values because they are amix of both lantitude and longitude into a a data type called point they are geographical coordinates
```

```python
df['Electric Utility'] = df['Electric Utility'].fillna('Addresses Outside Of Washington')
```

```python
df['Base MSRP'] = df['Base MSRP'].replace(0,np.nan)
df['Base MSRP'].fillna(df['Base MSRP'].median(),inplace = True)
```

```python
df['Electric Range'] = df['Electric Range'].replace(0,np.nan)
df['Electric Range'].fillna(df['Electric Range'].median(),inplace = True)
```

```python
df.dropna(subset=['Base MSRP','Electric Range'], inplace=True)
```
